### Reshaping, Stacking, Squeezing, and Unsqueezing tensors

````python
X = torch.arange(1, 10, dtype=torch.float)
````

````python
X
````

````python
X.reshape(3, 3)
````

````python
X.reshape(9, 1)
````

````python
X.view(1, 3, 3) # view of a tensor shares the same memory
                # with the input, so if I allocate X.view(1, 3,
                # 3) to a new variable Z. If I make a change to
                # Z, X will also change. Reshape doesn't do that!
````

````python
torch.stack([X, X], dim=0)
````

````python
torch.stack([X, X], dim=1)
````

````python
Z = X.reshape(1, 3, 3)
````

````python
torch.stack([Z, Z], dim=2)
````

````python
torch.hstack([X, X])
````

````python
torch.vstack([X, X])
````

````python
C = X.reshape([1, 9])
````

````python
C.shape
````

````python
squeezed_C = C.squeeze()
````

````python
squeezed_C.shape
````

#### Reproducibility

````python
A = torch.rand(3, 4)
B = torch.rand(3, 4)

print(A, '\n', B, '\n', A == B)
````

````python
# set random seed
RANDOM_SEED = 42

# NOTE: The seed has to be set after every random function call
````

````python
torch.manual_seed(RANDOM_SEED)
A = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)
B = torch.rand(3, 4)

print(A, '\n', B, '\n', A == B)
````
